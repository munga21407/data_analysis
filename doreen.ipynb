```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import yfinance as yf

yf.pdr_override()

data = pdr.get_data_yahoo("MSFT", "2010-01-01", "2023-07-14")

x = data.index
y_close = data['Close']
y_high = data['High']
y_low = data['Low']

plt.figure(figsize=(10, 6))
plt.plot(x, y_close, label='Close')
plt.plot(x, y_high, label='High')
plt.plot(x, y_low, label='Low')

plt.title('Microsoft Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.show()


```

```python
import pandas as pd
import plotly.graph_objects as go
from pandas_datareader import data as pdr
import yfinance as yf

yf.pdr_override()

data = pdr.get_data_yahoo("MSFT", "2010-01-01", "2023-07-14")

fig = go.Figure()

fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name='Close'))
fig.add_trace(go.Scatter(x=data.index, y=data['High'], name='High'))
fig.add_trace(go.Scatter(x=data.index, y=data['Low'], name='Low'))

fig.update_layout(
    title='Microsoft Stock Prices',
    xaxis_title='Date',
    yaxis_title='Price',
    hovermode='x'
)

fig.show()
```
