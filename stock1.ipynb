```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('sep100.csv')
x = data.index
y_AAPL = data['AAPL']
y_VZ = data['VZ']
y_WBA= data['WBA']
y_WFC=data['WFC']
plt.figure(figsize=(10, 6))
plt.plot(x, y_AAPL, label='AAPL')
plt.plot(x, y_VZ, label='VZ')
plt.plot(x, y_WFC, label='WFC')
plt.plot(x, y_WBA, label='WBA')
plt.title('Open Price Stocks - All S&P100 trends💲')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

print(data)


```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('sep100.csv')

fig = px.line(df, x=df.index, y=['AAPL', 'VZ', 'WFC', 'WBA'],
              labels={'value':'Price'},
              title='Open Price Stocks - All S&P100 trends')

fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Price')

fig.show()
```
